```python
from random import randint
import numpy as np
from itertools import combinations, chain
```

```python
char_set_base = ["A", "B", "C", "D"]
char_set_special = ["X", "Y", "Z"]
char_set_rare = ["J", "K"]
char_set_all =  char_set_base+\
                char_set_special+\
                char_set_rare
```

```python
bag_of_og = char_set_base*16+\
            char_set_special*8+\
            char_set_rare*4
```

```python
def first_draw(spell):
    count = 0
    monte=10000
    for sim in range(monte):
        bag_of_tiles = list(np.copy(bag_of_og))
        hand = []
        for draw in range(7):
            val = bag_of_tiles[randint(0, len(bag_of_tiles)-1)] # draw a card
            hand.append(val)
            bag_of_tiles.remove(val) # remove from deck
            flag = True
        for char in spell:
            if char in hand:
                hand.remove(char)
            else:
                flag = False
                break
        if flag:
            count +=1

    #     print(hand)
    return round(count/monte*100, 2)


def run_sim(sl, func=first_draw):
    for i in range(len(sl)):
        print(sl[i], func(sl[i]))
```

```python
## no repeats
sl = [["A", "B"],
      ["A", "B", "C"],
      ["A", "B", "X"],
      ["A", "X", "Y"],
      ["A", "X", "J"],
      ["Y", "X", "J"],
      ["A", "B", "C", "D"],
      ["A", "B", "C", "X"],
      ["A", "B", "X", "Y"],
      ["A", "B", "X", "J"],
      ["A", "B", "C", "D", "X"],
      ["A", "B", "C", "Y", "X"],
      ["A", "B", "J", "Y", "X"],
      ["A", "B", "C", "D", "X", "Y"],
      ["A", "B", "C", "Z", "X", "Y"],
      ["A", "B", "C", "J", "X", "Y"],
      ["A", "B", "C", "D", "X", "Y", "Z"],
      ["A", "B", "C", "D", "X", "Y", "J"]
     ]
```

```python
run_sim(sl)
```

```python
## The generic combinations for 2 symbol spells
two_gen =["A", "X", "J", "K"]
```

```python
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)  # allows duplicate elements
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))

def get_all_combos(stuff):
    count = 0
    for i, combo in enumerate(powerset(stuff), 1):
        size = len(combo)
        if size==0 or size>7: continue
        if size==1 and combo[0]!="A": continue
        if size==2 and combo[1]!="B": continue
        if size==3 and combo[2]!="C": continue
        if size==4 and combo[3]!="D": continue
        if size==5 and combo[4]!="X": continue
        print('combo #{}: {}'.format(count, combo))
        count+=1
```

```python
get_all_combos(char_set_all)
```
